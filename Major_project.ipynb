# Displaying the percentage of outliers in each column containing continuous values

```python
dummy_dataframe=pd.DataFrame()
outlier_data_frame_columns,no_of_Outliers,p_of_Outliers=[],[],[]
for i in data_set.columns:
  if len(data_set[i].unique())>2:
    outlier_data_frame_columns.append(i)
    q1=data_set[i].quantile(0.25)
    q3=data_set[i].quantile(0.75)
    iqr=q3-q1
    lower_limit=q1-1.5*iqr
    upper_limit=q3+1.5*iqr
    def taking_outlier_values(data):
      if data<lower_limit or data>upper_limit:
        return data
      return 0
    dummy_dataframe['applied_values']=data_set[i].apply(taking_outlier_values)
    len_outlier_values=len(dummy_dataframe['applied_values'][dummy_dataframe['applied_values']!=0])
    p_of_outliers=int((len_outlier_values/len(data_set))*100)
    no_of_Outliers.append(len_outlier_values)
    p_of_Outliers.append(p_of_outliers)
outlier_data_frame=pd.DataFrame(index=[outlier_data_frame_columns])
outlier_data_frame['no_of_Outliers']=no_of_Outliers
outlier_data_frame['%_of_Outliers']=p_of_Outliers
outlier_data_frame





```

```python

def apply_color(val):
  font_color='black'
  background_color='white'
  if (val>=0.9 and val<1) or (val<=-0.9 and val>=-1):
    font_color='green'
    background_color='yellow'
  elif (val>=0.7 and val<0.9) or (val>-0.9 and val<=-0.7):
    font_color='green'
  elif (val>=0.4 and val<0.7) or (val>-0.7 and val<=-0.4):
    font_color='blue'
  elif val==1:
    font_color='black'
  else:
    font_color='red'
  return 'color: %s;background-color: %s' % (font_color,background_color)
```

# Multicollinearity of independent variables(correlation>=0.9 or correlation<=-0.9)

```python
multi_collinearity,multi_collinearity_values,corr_with_target={},{},{}
for c1 in data_set.columns[1:]:
  corr_with,corr_values,corr_value_target=[],[],0
  for c2 in data_set.columns[1:]:
    if c1!=c2 and (corr_data_set[c1][c2]>=0.9 or corr_data_set[c1][c2]<=-0.9):
      corr_with.append(c2)
      corr_values.append(round(corr_data_set[c1][c2],5))
      corr_value_target=round(corr_data_set['Col2'][c1],5)
      multi_collinearity[c1]=corr_with
      multi_collinearity_values[c1]=corr_values
      corr_with_target[c1]=corr_value_target
```

# Using StackingClassifier to build the model
* This is an **ensemble learning method** which is used **increase the performance of the model**.
* Took DecisionTreeClassifier,RandomForestClassifier,KNeighborsClassifier,Supportvectormachinesclassifier into stackmodel

```python
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.svm import SVC
svm=SVC()
lr=LogisticRegression()
clf=[('dtc',dtc),('rfc',rfc),('knn',knn),('svm',svm)] #Taking DecisionTreeClassifier,RandomForestClassifier,KNeighborsClassifier,Supportvectormachinesclassifier
stackmodel=StackingClassifier(estimators=clf,final_estimator=lr)
stackmodel.fit(train_data_x,train_data_y)
```

# Checking f1_score for StackingClassifier

```python
confusionmatrix=confusion_matrix(train_data_y,stackmodel.predict(train_data_x))
```

```python
precision=confusionmatrix[0][0]/(confusionmatrix[0][0]+confusionmatrix[0][1])
recall=confusionmatrix[0][0]/(confusionmatrix[0][0]+confusionmatrix[1][0])
f1_score_stacking=2*precision*recall/(precision+recall)
f1_score_stacking
```

# Accuracy for StackingClassifier

```python
Accuracy_stacking=(confusionmatrix[0][0]+confusionmatrix[1][1])/(confusionmatrix[0][0]+confusionmatrix[1][1]+confusionmatrix[0][1]+confusionmatrix[1][0])
Accuracy_stacking
```
